```python
countyp1=pd.read_csv('counties-page-1-table-1.csv')
countyp2=pd.read_csv('counties-page-2-table-1.csv',header=None)
countyp3=pd.read_csv('counties-page-3-table-1.csv',header=None)
```

```python
countyp2.columns=['county', 'population', 'total_units','occupied_units', 'vacant_units']
countyp3.columns=['county', 'population', 'total_units','occupied_units', 'vacant_units']
```

```python
census=pd.concat([countyp1,countyp2,countyp3],ignore_index=True)
```

```python
census
```

```python
census['county']=census['county'].str.replace('County', '', regex=True)
```

```python
census
```

```python
df=census.replace(',','',regex=True).copy()
```

```python
df
```

```python
df.info()
```

```python
df1=df.drop(['total_units','occupied_units', 'vacant_units'],axis=1)
```

```python
df1
```

```python
df1['population']=df1['population'].astype(int)
```

```python
df1['county_id']=np.arange(0, 99)
```

```python
df1.insert(0, 'county_id', df1.pop('county_id'))
```
